```python
list_split_tweets = []
for entry in df['content'] :       
        entry = entry.translate({ ord(c): " "+c+" " for c in "!@#$%^&*()[]{}';:,./?\|`~-=_+" }) 
        split_tweet = entry
        list_split_tweets.append(split_tweet.split(" "))           
```

```python
import numpy as np
embeddings_text = open('glove50.txt')
WordToVec       = dict()
for line in embeddings_text:
    entry_array = line.split(" ")
    word        = entry_array[0] 
    word_vector = np.array(entry_array[1:],dtype='float32')
    WordToVec[word] = word_vector
```

```python
WordToVec
```

```python
list_sentences = []
for split_tweet in list_split_tweets:
    list_vec = []
    for word in split_tweet:
        try:    
            list_vec.append(WordToVec[word.lower()])
        except Exception as e:
            print(e)
    senTomatrix = np.array(list_vec)
    list_sentences.append(senTomatrix)
```
